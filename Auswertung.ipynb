```python
import matplotlib.pyplot as plt
import csv
import numpy as np

linethickness=1.0
plt.rc('figure', figsize=(20, 10))

with open ("LogData.csv") as data:
    reader=list(csv.reader(data,delimiter=','))

logdata=np.asarray(reader,dtype=float)

mean=np.mean(logdata[:,1])
#plt.plot(logdata[:,0],logdata[:,1],linewidth=linethickness,label="Winkel Beschleunigungssensor")
#plt.plot(logdata[:,0],logdata[:,12],label="Gyroskop")
plt.plot(logdata[:,0],logdata[:,2],linewidth=linethickness,label="Gefilterter Winkel")
plt.plot(logdata[:,0],logdata[:,3],linewidth=linethickness,label="Zielwinkel")
fig=plt.gca()

plt.ylabel("Winkel in Grad")
plt.xlabel("Zeit in s")
plt.legend()
plt.grid(visible=True)
plt.hlines(mean,0,logdata[-1,0],linewidth=2)
plt.text(logdata[-1000,0],mean+2,"Mittelwert: "+str(mean))
#plt.xlim([2,10])
#plt.ylim([-5,5])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(logdata[:,0],logdata[:,8],label="Abweichung Gyroskop ungefiltert [°]]")
plt.legend()
plt.subplot(2,2,2)
plt.plot(logdata[:,0],logdata[:,9],label="Gyro [°/sec]")
plt.legend()
plt.subplot(2,2,3)
plt.plot(logdata[:,0],logdata[:,10],label="Abweichung Gyroskop gefiltert [°]",color='r')
plt.legend()
plt.subplot(2,2,4)
plt.plot(logdata[:,0],logdata[:,11],label="delta T")
plt.legend()
plt.show()
```
