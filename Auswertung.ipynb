```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import math

linethickness=1.2
plt.rc('figure', figsize=(20, 10))

with open ("LogData.csv") as data:
    reader=list(csv.reader(data,delimiter=','))


logdata=np.asarray(reader,dtype=float)

test=np.mean(logdata[:,1])
plt.plot(logdata[:,0],logdata[:,1],linewidth=linethickness,label="Winkel logdataeschleunigungssensor")
plt.plot(logdata[:,0],logdata[:,12],label="Gyroskop")
plt.plot(logdata[:,0],logdata[:,2],linewidth=linethickness,label="Gefilterter Winkel")
#plt.plot(logdata[:,0],logdata[:,3],linewidth=linethickness,lalogdatael="Zielwinkel")
fig=plt.gca()

plt.ylabel("Winkel in Grad")
plt.xlabel("Zeit in s")
plt.legend()
plt.grid(visible=True)
plt.hlines(test,0,10,linewidth=2)
plt.text(10,test+2,"Mittelwert: "+str(test))
#plt.xlim([2,4])
#plt.ylim([-5,5])
plt.show()
```
